```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import os
%matplotlib inline
```

```python
ds = pd.read_csv('./train.csv')
print ds.head(n=5)
print ds.shape
data = ds.values## .values command converts ds into a numpy array
print data.shape
```

```python
y_train = data[:, 0]## first column i.e the labels
X_train = data[:, 1:]##everything second column onwards
plt.imshow(X_train[3].reshape((28, 28)), cmap='gray')##printing image 3
plt.show()
X_mean_img=X_train.mean(axis=0)## mean image 
plt.imshow(X_mean_img.reshape((28, 28)), cmap='gray')## reshaping 784 pixels of mean image into 28x28 image
plt.show()
```

```python
X_train = (X_train - X_train.mean(axis=0))/(X_train.std(axis=0) + 1e-03)##normalizing the data
plt.figure(0)
plt.imshow(X_train[3].reshape((28, 28)), cmap='gray')##printing image 3
plt.show()
print X_train.shape
print X_train.shape[0]
```

```python
def dist1(x1, x2):
    return np.sqrt(((x1 - x2)**2).sum())

def dist2(x1, x2):
    return abs(x1-x2).sum()

def dist3(x1, x2):
    return np.sqrt((0.5*((np.sqrt(abs(x1)) - np.sqrt(abs(x2)))**2).sum()))




def knn(X_train, x, y_train, k=5):## x is the chosen example whose distance is calc with all entries in X_train
    vals = []
    
    for i in range(X_train.shape[0]):
        v = [dist1(x, X_train[i, :]), y_train[i]]
        vals.append(v)###vals ia a list whos each entry is of the form [distance b/w x and ith training example,label[i]]
    
    updated_vals = sorted(vals, key=lambda x: x[0])#sort on basis of first entry of vals i.e distances
    pred_arr1 = np.asarray(updated_vals[:k])##taking first k examples
    pred_arr = np.unique(pred_arr1[:, 1], return_counts=True)
    ##pred_arr is a tuple 
    ##first elementof tuple i.e pred_arr[0] is an array of unique labels in k examples
    ##second element of tuple i.e pred_arr[1] is no. of times the corrosponding label(in pred_arr[0]) appears 
    
    j = pred_arr[1].argmax()##returns index of higher no. in pred_arr[1]
                               #pred_arr[0][j] is the predicted label    i.e (jth element of array pred_arr[0])
    return pred_arr1,pred_arr,j, pred_arr[0][j]


def knn1(X_train, x, y_train, k=5):## x is the chosen example whose distance is calc with all entries in X_train
    vals = []
    
    for i in range(X_train.shape[0]):
        v = [dist2(x, X_train[i, :]), y_train[i]]
        vals.append(v)###vals ia a list whos each entry is of the form [distance b/w x and ith training example,label[i]]
    
    updated_vals = sorted(vals, key=lambda x: x[0])#sort on basis of first entry of vals i.e distances
    pred_arr1 = np.asarray(updated_vals[:k])##taking first k examples
    pred_arr = np.unique(pred_arr1[:, 1], return_counts=True)
    ##pred_arr is a tuple 
    ##first elementof tuple i.e pred_arr[0] is an array of unique labels in k examples
    ##second element of tuple i.e pred_arr[1] is no. of times that label appears in k examples
    j = pred_arr[1].argmax()##the index(in pred_arr[0]) of the label that occurs max times in pred_arr[0]
                               #pred_arr[0][pred] is the predicted label    i.e (jth element of array pred_arr[0])
    return pred_arr1,pred_arr,j, pred_arr[0][j]


def knn2(X_train, x, y_train, k=5):## x is the chosen example whose distance is calc with all entries in X_train
    vals = []
    
    for i in range(X_train.shape[0]):
        v = [dist3(x, X_train[i, :]), y_train[i]]
        vals.append(v)###vals ia a list whos each entry is of the form [distance b/w x and ith training example,label[i]]
    
    updated_vals = sorted(vals, key=lambda x: x[0])#sort on basis of first entry of vals i.e distances
    pred_arr1 = np.asarray(updated_vals[:k])##taking first k examples
    pred_arr = np.unique(pred_arr1[:, 1], return_counts=True)
    ##pred_arr is a tuple 
    ##first elementof tuple i.e pred_arr[0] is an array of unique labels in k examples
    ##second element of tuple i.e pred_arr[1] is no. of times that label appears in k examples
    j = pred_arr[1].argmax()##the index(in pred_arr[0]) of the label that occurs max times in pred_arr[0]
                               #pred_arr[0][pred] is the predicted label    i.e (jth element of array pred_arr[0])
    return pred_arr1,pred_arr,j, pred_arr[0][j]
```

```python

a = int(np.random.random() * X_train.shape[0])## gives an integer between 0 and X_train.shape[0]
##random.random() generates float number b/w 0 and 1
##int converts the no. into and integer
x = X_train[a]## ath image

q,r,s,t= knn2(X_train[:10000], x, y_train[:10000], k=7)
print q,r,s,t
print y_train[a]##correct label

plt.figure(0)
plt.imshow(x.reshape((28, 28)), cmap='gray')
plt.show()

```
